```python
#Get and load the data

import numpy as np

sepal_length, sepal_width, petal_length, petal_width = np.genfromtxt('iris.csv', delimiter=",", usecols=(0,1,2,3), unpack=True, dtype=float)    	
species = np.genfromtxt('iris.csv', delimiter=",", usecols=(4), unpack=True, dtype=str)

for i in range(len(sepal_length)):
    print(sepal_length[i], sepal_width[i], petal_length[i], petal_width[i], species[i])
```

```python
#Create a simple plot

import matplotlib.pyplot as pl

pl.rcParams['figure.figsize'] = (14, 6)
pl.scatter(sepal_length, sepal_width)
pl.title('Sepal Length vs Width', fontsize=14)
pl.xlabel('Sepal Length')
pl.ylabel('Sepal Width')
pl.show()
```

```python
#Create a more complex plot

import matplotlib.pyplot as pl

pl.rcParams['figure.figsize'] = (14, 6)

colours = {'setosa': 'r', 'versicolor': 'g', 'virginica': 'b'}

pl.scatter(sepal_length, sepal_width, c=[colours[i] for i in species], label=[colours[i] for i in colours])
pl.title('Sepal Length vs Width', fontsize=14)
pl.xlabel('Sepal Length')
pl.ylabel('Sepal Width')
pl.show()
```

```python
#Use seaborn

import seaborn as sns
sns.set(style="ticks")

df = sns.load_dataset("iris")

%matplotlib inline
sns.pairplot(df, hue="species")
```

```python
#Fit a line

pl.rcParams['figure.figsize'] = (14.0, 8.0)

x = np.linspace(0,8,1000)
bestLine = np.polyfit(petal_length, petal_width, 1)
y = bestLine[0] * x + bestLine[1]

pl.scatter(petal_length, petal_width)
pl.plot(x,y,'g')
pl.xlabel('Petal Length')
pl.ylabel('Petal Width')
pl.show()
```

```python
#Calculate the R-squared value

np.corrcoef(petal_length, petal_width)[0][1]**2
```
